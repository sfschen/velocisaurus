```python
from matplotlib.lines import Line2D

# Function to extract components
def extract_components(bf, Nk):
    return {
        'p0': bf[0:Nk],
        'p2': bf[Nk:2*Nk],
        'p4': bf[2*Nk:3*Nk],
        'v1': -bf[3*Nk:4*Nk],
        'v3': -3*bf[4*Nk:5*Nk],
        's0': -bf[5*Nk:6*Nk],
        's2': -bf[6*Nk:7*Nk]
    }

# Extract theory
thy_ept = extract_components(bf, Nk)

# Set up figure
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Panel definitions: [(data_keys, theory_keys, colors, title)]
panels = [
    (['p0','p2'],[r'$P_0$', r'$P_2$', r'$P_4$'], ['C0', 'C1', 'C2']),
    (['v1'],[r'$v_1$', r'$3v_3$'],  ['C0', 'C1']),
    (['s0','s2'],[r'$\sigma^2_0$', r'$\sigma^2_2$'],  ['C0', 'C1'])
]

ylims = [(-50, 1500), (0, 980), (-500, 200)]  # y-limits per panel
kpows = [1, 2, 3]

# Data dict for easy access
data_dict = {
    'p0': (p0dat, p0err),
    'p2': (p2dat, p2err),
    'p4': (p4dat, p4err),
    'v1': (-v1dat, v1err),
    'v3': (-3*v3dat, v3err),
    's0': (-s0dat, s0err),
    's2': (-s2dat, s2err)
}

# Plot each panel
for i, (data_keys, legends, colors) in enumerate(panels):

    ax = axes[i]
    
    # Plot data and theory
    for j, key in enumerate(data_keys):
        dat, err = data_dict[key]
        ax.errorbar(kdat, kdat**kpows[i] * dat, yerr=kdat**kpows[i] * err, fmt='o', ms=4, color=colors[j], label=legends[j])

        ax.plot(kdat, kdat**kpows[i] * thy_ept[key], color=colors[j], linestyle='-')

    ax.set_xlim(0.00, 0.2)
    ax.set_ylim(*ylims[i])
    #ax.set_title(title)
    ax.set_xlabel(r'$k$ [$h$ Mpc$^{-1}$]')
    if i == 0:
        ax.set_ylabel(r'$i^n k^{n+1} \Xi^{(n)}_\ell$ [$h^{-2}$ Mpc$^2$]')

    if i != 2:
        ax.legend(ncol=4, loc='upper right')
    else:
        # Get existing legend handles/labels
        handles, labels = ax.get_legend_handles_labels()

        # Add EPT/LPT style indicators
        handles += [
        Line2D([0], [0], color='k', linestyle='-', label='EPT'),
        Line2D([0], [0], color='k', linestyle='--', label='LPT')
        ]
        labels += ['EPT', 'LPT']

        ax.legend(handles, labels, ncol=4, loc='upper right')

plt.tight_layout()
plt.show()


#plt.savefig('Figures/fits_z_0p1_logM_12p5_13p0.pdf')
```
